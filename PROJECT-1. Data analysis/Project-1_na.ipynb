```python
def get_expirience(arg):
    if arg == '':
        return np.nan
    elif arg == 'Не указано':
        return np.nan
    else:
        exclude_list = ['лет', 'год', 'года']
        expirience_list = arg.split(' ')
        short_list = expirience_list[:6]
        print(short_list,'\n',expirience_list)
        for i in range(len(short_list)):
            if short_list[i] in exclude_list:
                expirience = short_list[i-1]
        return int(expirience)
get_expirience(1150)
```
